```python
import pandas as pd
import json

# 1. Load the JSON file that Team Member 1 generated
with open('data/historical_ads.json', 'r', encoding='utf-8') as f:
    data = json.load(f)

# 2. Convert to a Pandas DataFrame
df = pd.DataFrame(data)

# 3. Look at the first 3 rows to see what the API gave us
display(df.head(3))

print("Here are all the keys (columns) from the API:")
print(df.columns.tolist())
```

```python
# Look at the keys you found in Cell 2. If the price column is called something else, change 'price' below.
price_key = 'price' 

def extract_price(price_data):
    if pd.isna(price_data): 
        return None
    # If the price is nested like {'value': 4500}
    if isinstance(price_data, dict) and 'value' in price_data:
        return price_data['value']
    # If it is already a simple number
    elif isinstance(price_data, (int, float)):
        return price_data
    return None

# Create a new column with ONLY clean numbers
df['clean_price'] = df[price_key].apply(extract_price)

# Show the cleaned prices
print(df[['title', 'clean_price']].head(5)
```

```python
# Remove any ads that don't have a price (e.g. "Bytes", "Bortskänkes")
valid_prices = df['clean_price'].dropna()

mean_price = valid_prices.mean()
min_price = valid_prices.min()
max_price = valid_prices.max()

print(f"📊 DATA SUMMARY FOR 'iPhone 13':")
print(f"Average Price: {mean_price:.0f} SEK")
print(f"Lowest Price:  {min_price:.0f} SEK")
print(f"Highest Price: {max_price:.0f} SEK")
```
